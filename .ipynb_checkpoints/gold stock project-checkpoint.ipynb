```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('gold_stock.csv')
print(df.head(5))
```

As we see the dataset has 6 columns showing the date, the closing value of gold stock in each day, the volume of gold stocks, the opeing price, the highest and lowest value for each day. this dataset contains the information of 10 years, starting from 2014 till 2024.
to check the higest and lowest price 

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
unique_years = df['Year'].unique()
unique_years
```

Now we are going to check the highest and lowest price of gold stocks for each year.

```python
df
```

```python

df_yearly_high_low = df.groupby('Year')['High'].agg([('Highest', 'max'), ('Lowest', 'min')])


print(df_yearly_high_low)
```

```python
mean = df.groupby('Year')['High'].agg('mean')
print('the average value of gold stock for each year is')
print(mean)
      
```

```python
std = df.groupby('Year')['High'].agg('std')
std
```

Our analysis of gold stock prices for each year reveals varying levels of variability. In 2016 and 2020, prices exhibited considerable fluctuations, with standard deviations of approximately 74.19 and 142.01, respectively. Conversely, 2017 saw relatively stable prices, with a standard deviation of approximately 35.21. Understanding these fluctuations is crucial for investors to assess risk and make informed decisions.


```python
plt.figure(figsize = (8,5))
mean_high_prices = df.groupby('Year')['High'].mean()
years = mean_high_prices.index
mean_low_prices = df.groupby('Year')['Low'].mean()
plt.plot(years, mean_high_prices, mean_low_prices)
# plt.plot(years, mean_low_prices, color = 'red')
plt.xlabel('Year')
plt.xticks(years, ['2014','2015','2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023'])
plt.ylabel('Average Price')
plt.title('Average value of high and low Price by Year')
plt.show()


```

## Gold Stock Market Trends

Upon examining the trends of the Gold stock market over the past ten years, it becomes evident that there has been a consistent upward trajectory. This trend implies promising investment opportunities for those looking to capitalize on the market's growth potential.

However, it's crucial to note that recent years have witnessed a notable increase in volatility within the Gold stock market. This volatility is highlighted by the widening gap between the highest and lowest prices observed during this period. While this heightened volatility introduces both opportunities and risks for investors, it underscores the importance of vigilance and informed decision-making when navigating the market.

In light of these findings, investors should approach investment opportunities in the Gold stock market with a keen awareness of the potential for both growth and volatility. By staying informed and adopting prudent risk management strategies, investors can effectively capitalize on the market's upward trend while mitigating the associated risks.

now lets see the seasonal trends in gold stock prices over the past 4 years

```python

df['Month'] = df['Date'].dt.month

# Group by month and calculate average 'Close' price
monthly_avg_close = df.groupby(['Year', 'Month'])['Close'].mean().reset_index()

# Plot seasonal trends
plt.figure(figsize=(10, 7))
for year in monthly_avg_close['Year'].unique():
    year_data = monthly_avg_close[monthly_avg_close['Year'] == year]
    plt.plot(year_data['Month'], year_data['Close'], label=year)

plt.title('Seasonal Trends in Gold Stock Prices')
plt.xlabel('Month')
plt.ylabel('Average Close Price')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True)
plt.show()
```
